**Regresión lineal múltiple**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Francia_limpio.csv')
df.head()
```

```python
df.info()
```

**PRIMER MODELO: host_acceptance_rate**

```python
from turtle import color

#sns.scatterplot(x='host_total_listings_count', y='host_acceptance_rate', color="blue", data=df)
sns.scatterplot(x='accommodates', y='host_acceptance_rate', color="red", data=df)
sns.scatterplot(x='bedrooms', y='host_acceptance_rate', color="yellow", data=df)
#sns.scatterplot(x='price', y='host_acceptance_rate', color="pink", data=df)
sns.scatterplot(x='reviews_per_month', y='host_acceptance_rate', color="green", data=df)
```

```python
#Declaramos las variables dependientes e independientes para la regresión lineal
Vars_Indep1= df[['accommodates', 'bedrooms','reviews_per_month']] #tengo múltiplas variables independientes
Var_Dep1= df['host_acceptance_rate']
```

```python
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
#Verificamos la función relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep1, y=Var_Dep1)
model.__dict__
```

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinación
model.score(Vars_Indep1,Var_Dep1)
```
